```python
import pandas as pd
```

```python
df = pd.read_csv(
    "FoodBalanceSheets_E_Africa_NOFLAG.csv", encoding='latin-1')
```

```python
df.head()
```

```python
# total sum of Animal Fat produced in 2014 and 2017
item_group = df.groupby('Item').sum()
item_2014_and_2017 = item_group[['Y2014', 'Y2017']]
item_2014_and_2017.loc['Animal fats']
```

```python
# mean and standard deviation across the dataset for year 2015
df['Y2015'].describe()
```

```python
# total number of missing data in 2016
missing_data_2016 = df['Y2016'].isnull().sum()
print(missing_data_2016)
Percentage_missing_data_2016 = missing_data_2016 / len(df['Y2016'])
print((Percentage_missing_data_2016 * 100).round(2))
```

```python
# highest correlation with Element Code
df[['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']].corrwith(df['Element Code'])
```

```python
# year with highest sum of Import Quantity
element_group = df.groupby('Element').sum()
element_by_year = element_group[['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']]
element_by_year.loc['Import Quantity']
# year 2017 has the highest import quantity
```

```python
# total number of the sum of Production in 2014
production_2014 = element_group['Y2014'].loc['Production']
production_2014
```

```python
# highest sum in 2018
df_2018_and_element = df[['Y2018', 'Element']].groupby('Element').sum()
df_2018_and_element.sort_values('Y2018', ascending=False)
# Domestic supply quantity has the highest sum in 2018
# Protein supply quantity (g/capita/day) has the 3rd lowest sum in 2018
```

```python
# total Import Quantity in Algeria in 2018
element_by_area = df[['Area', 'Y2018', 'Element']].groupby(['Element', 'Area']).sum()
element_by_area.loc['Import Quantity']
# total import in Algeria in 2018 is 36238.29
```
